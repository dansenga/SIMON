```python
from transformers import MarianMTModel, MarianTokenizer

model_name = "Helsinki-NLP/opus-mt-en-fr"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
text = "I am studying artificial intelligence at the university."

inputs = tokenizer(text, return_tensors="pt", padding=True)
translated = model.generate(**inputs)

result = tokenizer.decode(translated[0], skip_special_tokens=True)
print(result)
```

```python
from transformers import MarianMTModel, MarianTokenizer
import whisper

model_whisper = whisper.load_model("small")
result = model_whisper.transcribe("test2.mp3")
texte_transcrit = result["text"]
langue_detectee = result["language"]
print("Texte transcrit :", texte_transcrit)
print("Langue détectée :", langue_detectee)

langue_sortie = "fr"

model_name = f"Helsinki-NLP/opus-mt-{langue_detectee}-{langue_sortie}"
tokenizer = MarianTokenizer.from_pretrained(model_name)
model_trad = MarianMTModel.from_pretrained(model_name)

inputs = tokenizer(texte_transcrit, return_tensors="pt", padding=True)
translated = model_trad.generate(**inputs)
texte_traduit = tokenizer.decode(translated[0], skip_special_tokens=True)
print("Texte traduit :", texte_traduit)
```
